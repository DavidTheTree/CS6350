```python
from sklearn.preprocessing import LabelEncoder
# Assign column names (manually or based on dataset structure)
df1.columns = ['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5', 'Feature6', 'Target']
df2.columns = ['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5', 'Feature6', 'Target']
# Initialize the label encoder
label_encoder = LabelEncoder()

# Apply label encoder to all columns in the DataFrame
for column in df1.columns:
    df1[column] = label_encoder.fit_transform(df1[column])

for column in df2.columns:
    df2[column] = label_encoder.fit_transform(df2[column])

# Display the first few rows of the transformed DataFrame
print(df1.head())
print(df2.head())
```

```python
#Create ID3DecisionTree
class ID3DecisionTree:
    def __init__(self, max_depth=None, metric='information_gain'):
        self.max_depth = max_depth
        self.metric = metric
        self.tree = None

    def entropy(self, y):
        """Calculate entropy of a dataset"""
        freq = np.bincount(y)
        probabilities = freq / len(y)
        return -np.sum([p * np.log2(p) for p in probabilities if p > 0])

    def majority_error(self, y):
        """Calculate majority error of a dataset"""
        freq = np.bincount(y)
        return 1 - np.max(freq) / len(y)

    def gini_index(self, y):
        """Calculate gini index of a dataset"""
        freq = np.bincount(y)
        probabilities = freq / len(y)
        return 1 - np.sum([p ** 2 for p in probabilities])

    def information_gain(self, X, y, feature):
        """Calculate information gain for a specific feature"""
        total_entropy = self.entropy(y)
        values, counts = np.unique(X[:, feature], return_counts=True)
        weighted_entropy = np.sum([
            (counts[i] / np.sum(counts)) * self.entropy(y[X[:, feature] == v])
            for i, v in enumerate(values)
        ])
        return total_entropy - weighted_entropy

    def split_data(self, X, y, feature):
        """Split the data based on a given feature"""
        values = np.unique(X[:, feature])
        splits = {}
        for value in values:
            splits[value] = (X[X[:, feature] == value], y[X[:, feature] == value])
        return splits

    def best_feature(self, X, y):
        """Select the best feature based on the chosen metric"""
        num_features = X.shape[1]
        if self.metric == 'information_gain':
            gains = [self.information_gain(X, y, i) for i in range(num_features)]
        elif self.metric == 'gini_index':
            gains = [-self.gini_index(y[X[:, i] == val]) for i in range(num_features) for val in np.unique(X[:, i])]
        elif self.metric == 'majority_error':
            gains = [-self.majority_error(y[X[:, i] == val]) for i in range(num_features) for val in np.unique(X[:, i])]
        return np.argmax(gains)

    def majority_class(self, y):
        """Return the majority class"""
        return Counter(y).most_common(1)[0][0]

    def build_tree(self, X, y, depth=0):
        """Recursively build the decision tree"""
        if len(np.unique(y)) == 1:
            return y[0]  # Pure class, return label

        if len(X) == 0 or (self.max_depth is not None and depth >= self.max_depth):
            return self.majority_class(y)

        feature = self.best_feature(X, y)
        tree = {feature: {}}
        splits = self.split_data(X, y, feature)

        for feature_value, (X_sub, y_sub) in splits.items():
            if len(y_sub) == 0:
                tree[feature][feature_value] = self.majority_class(y)
            else:
                tree[feature][feature_value] = self.build_tree(X_sub, y_sub, depth + 1)

        return tree

    def fit(self, X, y):
        """Train the decision tree"""
        self.tree = self.build_tree(X, y)

    def predict_one(self, x, tree):
        """Predict for a single instance"""
        if not isinstance(tree, dict):
            return tree
        feature = list(tree.keys())[0]
        feature_value = x[feature]
        return self.predict_one(x, tree[feature].get(feature_value, self.majority_class([0])))

    def predict(self, X):
        """Predict for a set of instances"""
        return np.array([self.predict_one(x, self.tree) for x in X])
```

```python
#Run the Tree with different depth.
X = df1.iloc[:, :-1].values  # Features (all columns except the last one)
y = df1.iloc[:, -1].values   # Target (last column)

tree1 = ID3DecisionTree(max_depth=1, metric='information_gain')#
tree1.fit(X, y)
print("Trained Tree:")
print(tree1.tree)

tree2 = ID3DecisionTree(max_depth=2, metric='information_gain')
tree2.fit(X, y)
print("Trained Tree:")
print(tree2.tree)

tree3 = ID3DecisionTree(max_depth=3, metric='information_gain')
tree3.fit(X, y)
print("Trained Tree:")
print(tree3.tree)

tree4 = ID3DecisionTree(max_depth=4, metric='information_gain')
tree4.fit(X, y)
print("Trained Tree:")
print(tree4.tree)

tree5 = ID3DecisionTree(max_depth=5, metric='information_gain')
tree5.fit(X, y)
print("Trained Tree:")
print(tree5.tree)

tree6 = ID3DecisionTree(max_depth=6, metric='information_gain')
tree6.fit(X, y)
print("Trained Tree:")
print(tree6.tree)

```

```python
from sklearn.metrics import accuracy_score

# Function to calculate training error
def calculate_training_error(tree, X, y):
    y_pred = tree.predict(X)  # Predict on the training set
    accuracy = accuracy_score(y, y_pred)  # Calculate accuracy
    error = 1 - accuracy  # Training error is 1 - accuracy
    return error

tree1 = ID3DecisionTree(max_depth=1, metric='information_gain')
tree1.fit(X, y)
error1 = calculate_training_error(tree1, X, y)
print(f"Training Error for depth 1: {error1}")

# For tree with max depth = 2
tree2 = ID3DecisionTree(max_depth=2, metric='information_gain')
tree2.fit(X, y)
error2 = calculate_training_error(tree2, X, y)
print(f"Training Error for depth 2: {error2}")

# For tree with max depth = 3
tree3 = ID3DecisionTree(max_depth=3, metric='information_gain')
tree3.fit(X, y)
error3 = calculate_training_error(tree3, X, y)
print(f"Training Error for depth 3: {error3}")

# For tree with max depth = 4
tree4 = ID3DecisionTree(max_depth=4, metric='information_gain')
tree4.fit(X, y)
error4 = calculate_training_error(tree4, X, y)
print(f"Training Error for depth 4: {error4}")

# For tree with max depth = 5
tree5 = ID3DecisionTree(max_depth=5, metric='information_gain')
tree5.fit(X, y)
error5 = calculate_training_error(tree5, X, y)
print(f"Training Error for depth 5: {error5}")

# For tree with max depth = 6
tree6 = ID3DecisionTree(max_depth=6, metric='information_gain')
tree6.fit(X, y)
error6 = calculate_training_error(tree6, X, y)
print(f"Training Error for depth 6: {error6}")
```
